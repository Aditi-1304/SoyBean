```python
# STEP 1: Load the saved model
from tensorflow.keras.models import load_model
import numpy as np
from tensorflow.keras.preprocessing import image

# Load the model from the saved file
model = load_model('/content/drive/MyDrive/Models/soybean_leaf_filter_mobilenetv2.h5')

# STEP 2: Load and preprocess the image
img_path = '/content/drive/MyDrive/Testing/frogeye.jpeg'  # Path to the image you want to predict
img = image.load_img(img_path, target_size=(224, 224))  # Resize image to match input size of the model

# Convert image to numpy array
img_array = image.img_to_array(img)

# Expand dimensions to match the batch shape (1, 224, 224, 3)
img_array = np.expand_dims(img_array, axis=0)

# Normalize the image as we did during training
img_array = img_array / 255.0

# STEP 3: Make a prediction
prediction = model.predict(img_array)

# Optional: If you saved class labels earlier, load them and map prediction to a label
import json

with open('/content/drive/MyDrive/Models/class_labels.json', 'r') as f:
    labels = json.load(f)

# Get the predicted class index and the corresponding label
predicted_class = 1 if prediction[0] > 0.5 else 0
predicted_label = list(labels.keys())[list(labels.values()).index(predicted_class)]

print(f"Predicted label: {predicted_label}")
```
