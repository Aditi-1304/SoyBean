```python
import os
import cv2
import numpy as np

# Path to the folder containing images
input_folder = r"C:\Users\ASUS\Desktop\Dataset\Healthy"
output_folder = r"C:\Users\ASUS\Desktop\Dataset\Healthy_p"

# Create the output folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Define a function for preprocessing a single image


def preprocess_image(image_path):
    # Load the image
    img = cv2.imread(image_path)

    # Resize the image to a standard size (e.g., 224x224)
    img_resized = cv2.resize(img, (224, 224))

    # Normalize the image (scale pixel values to range [0, 1])
    img_normalized = img_resized / 255.0

    return img_normalized


# Loop through all images in the folder
for filename in os.listdir(input_folder):
    # Check if the file is an image (optional)
    if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff')):
        file_path = os.path.join(input_folder, filename)

        # Preprocess the image
        preprocessed_image = preprocess_image(file_path)

        # Save the preprocessed image to the output folder
        output_file_path = os.path.join(output_folder, filename)
        # Convert the image back to uint8 format for saving
        preprocessed_image_uint8 = (preprocessed_image * 255).astype(np.uint8)
        cv2.imwrite(output_file_path, preprocessed_image_uint8)

        print(f"Processed and saved: {filename}")

print("Batch processing completed.")
```
