```python
import os
import shutil

# Paths of the two folders
folder1 = '/content/drive/MyDrive/Sample/SoyBean'  # Replace with your first folder path
folder2 = '/content/drive/MyDrive/Sample/Augmented_SoyBean'  # Replace with your second folder path

# Output folder for combined images
output_folder = '/content/drive/MyDrive/Sample/Soybean'
os.makedirs(output_folder, exist_ok=True)

# Combine images from folder1
for img_file in os.listdir(folder1):
    img_path = os.path.join(folder1, img_file)

    # Ensure it's an image file
    if img_path.endswith(('.jpg', '.png', '.jpeg')):
        # Copy image to output folder
        shutil.copy(img_path, os.path.join(output_folder, img_file))

# Combine images from folder2
for img_file in os.listdir(folder2):
    img_path = os.path.join(folder2, img_file)

    # Ensure it's an image file
    if img_path.endswith(('.jpg', '.png', '.jpeg')):
        # Rename file if a file with the same name exists in the output folder
        base_name, ext = os.path.splitext(img_file)
        counter = 1
        new_name = img_file
        while os.path.exists(os.path.join(output_folder, new_name)):
            new_name = f"{base_name}_{counter}{ext}"
            counter += 1

        # Copy image to output folder
        shutil.copy(img_path, os.path.join(output_folder, new_name))

print(f"Images combined successfully into: {output_folder}")
```
