```python
from google.colab import drive
import os
import matplotlib.pyplot as plt
import cv2
import numpy as np
from collections import Counter

# Mount Google Drive
drive.mount('/content/drive')
```

```python
# Define the dataset path
DATASET_PATH = '/content/drive/My Drive/Data'

# Get class labels (folder names)
class_labels = sorted(os.listdir(DATASET_PATH))  # Sorted to maintain order
num_classes = len(class_labels)

# Count images in each class
class_counts = {label: len(os.listdir(os.path.join(DATASET_PATH, label))) for label in class_labels}

# Plot class distribution
plt.figure(figsize=(12, 6))
plt.bar(class_counts.keys(), class_counts.values(), color='skyblue')
plt.xlabel("Class")
plt.ylabel("Number of Images")
plt.title("Soybean Leaf Dataset Distribution")
plt.xticks(rotation=90)  # Rotate labels for better visibility
plt.show()

# Display sample images from each class (limit to 17)
num_samples = min(17, num_classes)
fig, axes = plt.subplots(3, 6, figsize=(18, 9))  # Create grid (adjustable)
axes = axes.flatten()

for i, label in enumerate(class_labels[:num_samples]):
    img_path = os.path.join(DATASET_PATH, label, os.listdir(os.path.join(DATASET_PATH, label))[0])
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    axes[i].imshow(img)
    axes[i].set_title(label)
    axes[i].axis("off")

# Hide extra subplots if there are fewer than 18 classes
for i in range(num_samples, len(axes)):
    axes[i].axis("off")

plt.tight_layout()
plt.show()
```
